```python
import numpy as np

import CAS.dmrghandler.src.dmrghandler as dmrghandler
from CAS.dmrghandler.src.dmrghandler import dmrg_calc_prepare
from pathlib import Path
import scipy
import pandas as pd
import os
import itertools
import xlsxwriter
```

## Load the MF tensors

```python
fcidump_path = Path("fcidumps_catalysts")
# Create Excel writer object
writer = pd.ExcelWriter("df_distance_20240610.xlsx", engine='xlsxwriter')

fcidump_df_path = "fcidumps_df"
```

## Load the Original tensors

```python
column_names = ["FCIDUMP file", "L2 norm difference for obt", "L2 norm difference for tbt", "F"]
data_to_be_added = []
fcidump_names = []
l2_norm_obt = []
l2_norm_tbt = []
f_list = []
for filename in os.listdir(fcidump_path):
    print(f"Current file: {filename}")
    (
    one_body_tensor,
    two_body_tensor,
    nuc_rep_energy,
    num_orbitals,
    num_spin_orbitals,
    num_electrons,
    two_S,
    two_Sz,
    orb_sym,
    extra_attributes,
    ) = dmrg_calc_prepare.load_tensors_from_fcidump(data_file_path=fcidump_path/Path(filename), molpro_orbsym_convention=True)
    
    
    (
    one_body_tensor_mf,
    two_body_tensor_mf,
    nuc_rep_energy,
    num_orbitals,
    num_spin_orbitals,
    num_electrons,
    two_S,
    two_Sz,
    orb_sym,
    extra_attributes,
    ) = dmrg_calc_prepare.load_tensors_from_fcidump(data_file_path=Path(fcidump_df_path)/Path(filename+"_tbt_combined_df"), molpro_orbsym_convention=True)
    
    
    l2_1 = scipy.linalg.norm(one_body_tensor - one_body_tensor_mf, ord=None)
    l2_2 = scipy.linalg.norm(two_body_tensor - two_body_tensor_mf, ord=None)
    F = l2_1 + l2_2
    
    fcidump_names.append(filename)
    l2_norm_obt.append(l2_1)
    l2_norm_tbt.append(l2_2)
    f_list.append(F)

data_to_be_added.append(fcidump_names)
data_to_be_added.append(l2_norm_obt)
data_to_be_added.append(l2_norm_tbt)
data_to_be_added.append(f_list)
```

```python
df_1 = pd.DataFrame()
#Iteratively add data to new columns
for column_name, column_data in zip(column_names, data_to_be_added):
    print(column_name)
    print(column_data)
    df_1[column_name] = column_data

df_1.to_excel(writer, sheet_name="MF distance result", index=False)
```
